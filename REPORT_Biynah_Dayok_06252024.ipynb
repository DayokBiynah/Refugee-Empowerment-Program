<h2>Let's create a KNN Classsifier model</h2>

```python
import sklearn.model_selection as model_selection
```

```python
X = df[['during_Life_Skills_Seminars', 'during_Employability_Services','during_Translation_and_Interpretation_Services', 'during_Legal_Services_Referral', 'demo_Gender',	'demo_Birth_Year',	'demo_Country_of_Origin']]
Y = df[['post_Became_a_US_Citizen']]
			
```

```python
X = pd.get_dummies(X, drop_first = True)
Y = pd.get_dummies(Y, drop_first = True)
X
```

```python
import sklearn.model_selection as model_selection
```

```python
splits = model_selection.train_test_split(X, Y, test_size = 0.2)
```

```python
import sklearn.neighbors as neighbors

import sklearn.metrics as metrics

import numpy as np

import sklearn.preprocessing as pp

import sklearn.pipeline as pipe
```

```python
std_clf = pipe.make_pipeline(pp.StandardScaler(), neighbors.KNeighborsClassifier(n_neighbors = 4))

std_clf
```

```python
std_clf.fit(splits[0], np.ravel(splits[2]))

```

```python
predictions = std_clf.predict(splits[1])

predictions
```

```python
import sklearn.metrics as metrics
metrics.confusion_matrix(splits[3],predictions)
```

```python
std_clf.predict_proba(splits[1])
```

<h2>Accuracy and metrics of our model</h2>

```python
metrics.accuracy_score(splits[3],predictions)
```

```python
print(metrics.classification_report(splits[3],predictions))
```

<h2>Let's build a logistic Regression</h2>

```python
import sklearn.model_selection as model_selection
```

```python
splits = model_selection.train_test_split(X,Y,test_size=0.2)
```

```python

import sklearn.linear_model as linear_model   
std_clf = pipe.make_pipeline(pp.StandardScaler(), linear_model.LogisticRegression())
```

```python
std_clf.fit(splits[0],np.ravel(splits[2]))
```

```python
predictions = std_clf.predict(splits[1])

predictions
```

<h2> Accuracy and other metrics of the model</h2>

```python
print('Accuracy:' + str(metrics.accuracy_score(splits[3],predictions)))
```

```python
print(metrics.classification_report(splits[3],predictions))
```

```python
probs = std_clf.predict_proba(splits[1])
rocMetrics = metrics.roc_curve(splits[3],probs[:,1])
fig = px.line(x=rocMetrics[0],y=rocMetrics[1])

fig.update_xaxes(title_text="Recall/True positive rate")

fig.update_yaxes(title_text="False positive rate")
```
